```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def lidl_funct():
    url = 'https://www.lidl.pt/c/produtos-lupilu/c471' # the url for the lidl baby products
    k = requests.get(url).text # the response
    
    soup = BeautifulSoup(k,'html.parser') # use the beautifulsoup to parse the html
    price_list_raw = soup.find_all(attrs ={'class':'pricebox__price'}) # get the raw list of the prices
    
    price_list = [] # the clean list of prices
    
    for element in price_list_raw:
        price_list.append((element.text).replace('\n','').strip()) # for loop to clean each item in the raw list
    
    
    soup = BeautifulSoup(k,'html.parser') # use beautifulsoup to parse the html
    name_list_raw = soup.find_all('h3', attrs= { 'class:','product__title'}) # the raw list of items names
    
    name_list = [] # the clean list of prices

    for element in name_list_raw: 
        name_list.append((element.text).replace('\n','').strip()) # the for loop for cleaning each item in the raw list
        
    new_dict = {
        'produtos': name_list,
        'preços': price_list,
        'supermercado': 'lidl'
    } # the dict that is going to get changed into a dataframe
    
    return pd.DataFrame(new_dict) # the return with a dataframe
    
    
```

```python
def continente_funct():
    url = 'https://www.continente.pt/bebe/' # the url for the continente baby products
    k = requests.get(url).text # the response

    soup = BeautifulSoup(k,'html.parser') # use the beautifulsoup to parse the html

    price_list_raw = soup.find_all(attrs ={'class':'ct-price-formatted'}) # get the raw list of the prices
    price_list = [] # the clean list of prices

    for element in price_list_raw:
        price_list.append((element.text).replace('\n','').strip().replace('€',''))

    soup = BeautifulSoup(k,'html.parser') # use beautifulsoup to parse the html

    name_list_raw = soup.find_all('a', attrs= {'class:','ct-tile--description'}) # the raw list of items names
    name_list = [] # the clean list of prices

    for element in name_list_raw:
        name_list.append(element.text)

    brand_list_raw = soup.find_all('p', attrs = {'class':'ct-tile--brand'})
    brand_list = []
    
    for element in brand_list_raw:
        brand_list.append(element.text)
        
    new_dict = {
        'produtos': name_list,
        'preços': price_list,
        'supermercado': 'continente',
        'marca': brand_list
    }
    
    return pd.DataFrame(new_dict) # the return with a dataframe
    
```

```python
continente = continente_funct()
```

```python
def marca(row):
    return (list(row.split(' ')))[0]
```

```python
def busca(string,df=geral):
    
    lenght = len(df['produtos']) # get the quantity of products to check
    
    # define the lists to check
    
    list_preços = [] 
    list_produtos = []
    list_marcas = []
    
    # for loop that checks first the correspondent elements, and then add the row to the lists

    for i in range(lenght):
        if string in df.iloc[i]['produtos']:
            list_produtos.append(df.iloc[i]['produtos'])
            list_preços.append(df.iloc[i]['preços'])
            list_marcas.append(df.iloc[i]['marca'])
    
    # dict that will be transformed into a dataframe
    
    busca_dict = {
        'produtos': list_produtos,
        'preços': list_preços,
        'marcas': list_marcas
    }
      
        
    return pd.DataFrame(busca_dict)
```
