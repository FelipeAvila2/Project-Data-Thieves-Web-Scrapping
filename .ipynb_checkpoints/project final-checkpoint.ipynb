```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def lidl_funct():
    url = 'https://www.lidl.pt/c/produtos-lupilu/c471' # the url for the lidl baby products
    k = requests.get(url).text # the response
    
    soup = BeautifulSoup(k,'html.parser') # use the beautifulsoup to parse the html
    price_list_raw = soup.find_all(attrs ={'class':'pricebox__price'}) # get the raw list of the prices
    
    price_list = [] # the clean list of prices
    
    for element in price_list_raw:
        price_list.append(float((element.text).replace('\n','').strip())) # for loop to clean each item in the raw list
    
    
    soup = BeautifulSoup(k,'html.parser') # use beautifulsoup to parse the html
    name_list_raw = soup.find_all('h3', attrs= { 'class:','product__title'}) # the raw list of items names
    
    name_list = [] # the clean list of prices

    for element in name_list_raw: 
        name_list.append((element.text).replace('\n','').strip()) # the for loop for cleaning each item in the raw list
        
    new_dict = {
        'produtos': name_list,
        'preços': price_list,
        'supermercado': 'lidl'
    } # the dict that is going to get changed into a dataframe
    
    return pd.DataFrame(new_dict) # the return with a dataframe
    
    
```

```python
def continente_funct():
    url = 'https://www.continente.pt/bebe/' # the url for the continente baby products
    k = requests.get(url).text # the response

    soup = BeautifulSoup(k,'html.parser') # use the beautifulsoup to parse the html

    price_list_raw = soup.find_all(attrs ={'class':'ct-price-formatted'}) # get the raw list of the prices
    price_list = [] # the clean list of prices

    for element in price_list_raw:
        price_list.append(float((element.text).replace('\n','').strip().replace('€','').replace(',','.')))

    soup = BeautifulSoup(k,'html.parser') # use beautifulsoup to parse the html

    name_list_raw = soup.find_all('a', attrs= {'class:','ct-tile--description'}) # the raw list of items names
    name_list = [] # the clean list of prices

    for element in name_list_raw:
        name_list.append(element.text)
        
    new_dict = {
        'produtos': name_list,
        'preços': price_list,
        'supermercado': 'continente'
    }
    
    return pd.DataFrame(new_dict) # the return with a dataframe
```

```python
def auchan_funct():
    url = 'https://www.auchan.pt/pt/beleza-higiene-e-bebe/' # the url for the lidl baby products
    k = requests.get(url).text # the response

    soup = BeautifulSoup(k,'html.parser') # use the beautifulsoup to parse the html
    price_list_raw = soup.find_all('span', attrs={'class':'value'}) # get the raw list of the prices

    price_list = [] # the clean list of prices

    for element in price_list_raw:
        if 'Price reduced from' in element.text:
            pass
        else:
            price_list.append(float((element.text).replace('\n','').replace('€','').replace(',','.').strip())) # for loop to clean each item in the raw list
    
    
    soup = BeautifulSoup(k,'html.parser') # use beautifulsoup to parse the html
    name_list_raw = soup.find_all('div', attrs={'class':"auc-product-tile__name"}) # the raw list of items names
    
    name_list = [] # the clean list of prices

    for element in name_list_raw: 
        name_list.append(((element.text).replace('\n','').strip()).capitalize()) # the for loop for cleaning each item in the raw list
        
    new_dict = {
        'produtos': name_list,
        'preços': price_list,
        'supermercado': 'Auchan'
    } # the dict that is going to get changed into a dataframe
    
    df = pd.DataFrame(new_dict)
    
    url = 'https://www.auchan.pt/pt/beleza-higiene-e-bebe/fraldas-e-toalhitas-de-bebe/' # the url for the lidl baby products
    k = requests.get(url).text # the response
    
    soup = BeautifulSoup(k,'html.parser') # use the beautifulsoup to parse the html
    price_list_raw1 = soup.find_all('span', attrs={'class':'value'}) # get the raw list of the prices

    price_list1 = [] # the clean list of prices

    for element in price_list_raw1:
        if 'Price reduced from' in element.text:
            pass
        else:
            price_list1.append(float((element.text).replace('\n','').replace('€','').replace(',','.').strip())) # for loop to clean each item in the raw list
    
    
    soup = BeautifulSoup(k,'html.parser') # use beautifulsoup to parse the html
    name_list_raw1 = soup.find_all('div', attrs={'class':"auc-product-tile__name"}) # the raw list of items names
    
    name_list1 = [] # the clean list of prices

    for element in name_list_raw1: 
        name_list1.append(((element.text).replace('\n','').strip()).capitalize()) # the for loop for cleaning each item in the raw list
        
    new_dict1 = {
        'produtos': name_list1,
        'preços': price_list1,
        'supermercado': 'Auchan'
    } # the dict that is going to get changed into a dataframe
    
    df1 = pd.DataFrame(new_dict1) # the second dataframe
    
    frame = [df, df1] # a list with the two dfs
     
    return pd.concat(frame) # concats the 2 dataframes and returns it
    
    
```

```python
auchan = auchan_funct()
```

```python
continente = continente_funct()
```

```python
lidl = lidl_funct()
```

```python
def busca(string,df=geral):
    
    lenght = len(df['produtos']) # get the quantity of products to check
    
    # define the lists to check
    
    list_preços = [] 
    list_produtos = []
    list_supermercado = []
    
    # for loop that checks first the correspondent elements, and then add the row to the lists

    for i in range(lenght):
        if string in df.iloc[i]['produtos']:
            list_produtos.append(df.iloc[i]['produtos'])
            list_preços.append(df.iloc[i]['preços'])
            list_supermercado.append(df.iloc[i]['supermercado'])       
    
    # dict that will be transformed into a dataframe
    
    busca_dict = {
        'produtos': list_produtos,
        'preços': list_preços,
        'supermercado': list_supermercado
    }
    
    df = (pd.DataFrame(busca_dict)).sort_values('preços')
    
    labels = ["Preço Ótimo!", "Preço baixo", "Preço normal", "Preço alto", "Preço muito alto!"]
    qbins = pd.cut(df['preços'],5, labels = labels)
    df['indicação'] = qbins
      
        
    return df
```
